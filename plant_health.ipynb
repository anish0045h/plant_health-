```python
numerical_cols = [
    'Soil_Moisture', 'Ambient_Temperature', 'Soil_Temperature',
    'Humidity', 'Light_Intensity', 'Soil_pH', 'Nitrogen_Level',
    'Phosphorus_Level', 'Potassium_Level', 'Chlorophyll_Content',
    'Electrochemical_Signal'
]
# Create a figure with a grid of subplots
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(20, 15))
fig.suptitle('Kernel Density Estimate (KDE) Plots for Numerical Columns', fontsize=20, y=1.02)
axes = axes.flatten()  # Flatten the 2D array of axes for easier iteration

for i, col in enumerate(numerical_cols):
    sns.kdeplot(df[col], fill=True, ax=axes[i], color='teal')
    axes[i].set_title(f'Distribution of {col}', fontsize=12)
    axes[i].set_xlabel(col, fontsize=10)
    axes[i].set_ylabel('Density', fontsize=10)

# Hide any empty subplots
for j in range(len(numerical_cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
custom_mapping = {'High Stress': 2, 'Moderate Stress': 1, 'Healthy': 0}
df['Plant_Health_Status_Encoded'] = df['Plant_Health_Status'].map(custom_mapping)

# Count unique values in Plant_Health_Status before encoding
unique_value_counts_before = df['Plant_Health_Status'].value_counts()

# Count unique values in Plant_Health_Status_Encoded after encoding
unique_value_counts_after = df['Plant_Health_Status_Encoded'].value_counts()
```

```python
data = df.drop(columns = ['Timestamp','Plant_ID','Plant_Health_Status'])
```

```python
data.head()
```

```python
x = data.drop(columns=['Plant_Health_Status_Encoded'])  # Features
y = data['Plant_Health_Status_Encoded']
```

```python
x.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train_scaled, y_train)
```

```python
dt_train_preds = decision_tree.predict(X_train_scaled)
dt_test_preds = decision_tree.predict(X_test_scaled)
dt_train_accuracy = accuracy_score(y_train, dt_train_preds)
dt_test_accuracy = accuracy_score(y_test, dt_test_preds)
```

```python
dt_train_accuracy
```

```python
dt_test_accuracy
```

```python
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train_scaled, y_train)
```

```python
rf_train_preds = rf_classifier.predict(X_train_scaled)
rf_test_preds = rf_classifier.predict(X_test_scaled)
rf_train_accuracy = accuracy_score(y_train, rf_train_preds)
rf_test_accuracy = accuracy_score(y_test, rf_test_preds)
```

```python
rf_train_accuracy
```

```python
rf_test_accuracy
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(X_train_scaled, y_train)
```

```python
knn_train_preds = knn_classifier.predict(X_train_scaled)
knn_test_preds = knn_classifier.predict(X_test_scaled)
knn_train_accuracy = accuracy_score(y_train, knn_train_preds)
knn_test_accuracy = accuracy_score(y_test, knn_test_preds)
```
